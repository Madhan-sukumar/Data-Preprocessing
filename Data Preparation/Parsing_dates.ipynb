```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
```

```python
#reading the dataset
df = pd.read_csv("./catalog.csv")
```

```python
df.head()
```

```python
#checking date column is in datetime format
print(df['date'].head())
```

```python
df['date'].dtype
```

```python
#parsing the 'object' date column into datetime
df['date_parsed'] = pd.to_datetime(df['date'],format="%m/%d/%y")
```

```python
df
```

```python
df['date_parsed'].head()
```

```python
#creating and getting the day of the month from date parsed column
df['day_of_month'] = df['date_parsed'].dt.day
day_of_month.head()
```

```python
df
```

```python
#removing the null values for plotting
day_of_month = day_of_month.dropna()
```

```python
#plotting the day of the month
sns.distplot(day_of_month,bins=31,kde=False)
```
